```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/Colab Notebooks/valorant-stats.csv")
df.head(10)
```

```python

sns.countplot(data=df, x='Weapon Type')
df['Weapon Type'].value_counts()
```

```python
df['Weapon Type'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
plt.hist(df['Price'])
```

```python
sns.distplot(df['Price'])
```

```python
sns.boxplot(df['Magazine Capacity'])
```

```python
df.describe()
```

```python
sns.scatterplot(x='Price', y='Fire Rate',hue='Wall Penetration', style='Weapon Type' ,data=df)
plt.show()
```

```python
sns.barplot(x='Fire Rate',y='Name',hue='Wall Penetration',data=df)
```

```python
sns.distplot(df[df['Wall Penetration']=='Low']['Price'],hist=False)
sns.distplot(df[df['Wall Penetration']=='Medium']['Price'],hist=False)
sns.distplot(df[df['Wall Penetration']=='High']['Price'],hist=False)
```

```python
sns.heatmap(pd.crosstab(df['Weapon Type'],df['Price']))
```

```python
sns.pairplot(df,hue='Wall Penetration')
```

```python
sns.lineplot(x='Price',y='BDMG_0',data=df)
```

```python
sns.heatmap(df.pivot(values='Fire Rate',index='Wall Penetration',columns='Name'))
```

```python
pivot_df = df.pivot(values='HDMG_0', index='HDMG_1', columns='Wall Penetration')
pivot_df = pivot_df.fillna(0)
sns.clustermap(pivot_df)
```
