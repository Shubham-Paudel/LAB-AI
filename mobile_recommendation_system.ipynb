```python
new_df=mobile[['mobile_name','tags']]
new_df['tags'] =new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
def Convert(string):
    li = string.lower()
    return li
```

```python
new_df['mobile_name']=new_df['mobile_name'].str.lower()
```

```python
new_df.head(20)
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()


print(vector)
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
  y =[]

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)


new_df['tags']=new_df['tags'].apply(stem)
print(new_df)
similarity = cosine_similarity(vector)
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
similarity[0]
```

```python

def recommend(name):
    name.lower() # make lowercase
    mobile_index = new_df[new_df["mobile_name"] == name].index[0]
    distances = similarity[mobile_index]
    mobile = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:7]
    for i in mobile:
        print(new_df.iloc[i[0]].mobile_name)
```

```python
recommend('samsung galaxy j3 pro (gold, 16 gb)')
```

```python
recommend('realme gt 2 pro (paper green, 256 gb)')
```

```python
recommend('oppo a15 (dynamic black, 32 gb)')
```
